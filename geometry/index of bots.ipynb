```python
pages_editors_graph = nx.read_gexf("data/pages-editors.gexf")
```

```python
bots = [ n for n in pages_editors_graph.nodes() if "bot" in n.lower() ]
```

```python
def table_to_html(data, cols=[]):
    html = "<table>"
    html += "<tr>"
    for column_content in cols:
        html += "<th>%s</th>" % (column_content)
    html += "</tr>"
    
    for d in data:
        html += "<tr>"
        for column_content in d:
            html += "<td>%s</td>" % (column_content)
        html += "</tr>"
    html += "</table>"
        
    return HTML(html)
```

```python
data = []

bots = sorted(bots, key=lambda b: -pages_editors_graph.node[b]["revisions"])

for b in bots:
    r  = pages_editors_graph.node[b]["revisions"]
    p  = len(pages_editors_graph[b])
    rp = float(r)/float(p)
    data.append([u"<a href=\"http://en.wikipedia.org/wiki/User:{0}\" target=\"_blank\">{0}</a>".format(b.split(":")[1]),
                r,
                p,
                rp])

display(table_to_html(data, ["bot name", "revisions", "edited pages", "revisions/page"]))
```
